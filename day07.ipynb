```python
from functools import cmp_to_key
from collections import Counter

inputs = open('inputs/day07.txt').readlines()
hands, bids = list(), list()

for line in inputs: 
    h, b = line.split()
    hands.append((h, b))

def strongest_suit(hand1, hand2, card_order): 
    for char_index in range(5): 
        c1 = hand1[0][char_index]
        c2 = hand2[0][char_index]

        c1 = card_order.index(c1)
        c2 = card_order.index(c2)

        if c1 < c2: 
            return 1
        elif c2 < c1: 
            return -1
         
    return 0 # Hands are equally strong...
def part1_compare_hands(hand1, hand2): 
    """
    Return 1 if hand1 is stronger than hand2.
     
    Return -1 if hand2 is stronger than hand1
    """
    cards_ordered = "AKQJT987654321"
    h1c = Counter(hand1[0])
    h2c = Counter(hand2[0])
    
    counts1 = sorted(list(h1c.values()), reverse=True)
    counts2 = sorted(list(h2c.values()), reverse=True)
    
    for c1, c2 in zip(counts1, counts2): 
        if c1 > c2: 
            return 1
        elif c2 > c1: 
            return -1
    
    # No returns? Let's return based on the first card!
    return strongest_suit(hand1, hand2, cards_ordered)

def get_joker_counts(hand_string): 
    joker_count = hand_string.count('J')
    hand_without_jokers = [x for x in hand_string if x != 'J']

    if hand_without_jokers: 
        counter = Counter(hand_without_jokers)
        counts = sorted(list(counter.values()), reverse=True)
        counts[0] += joker_count 
    else: # Only jokers 
        counts = [5]
    return counts

def part2_compare_hands(hand1, hand2): 
    cards_ordered = "AKQT98765432J"

    # Remove the jokers...

    counts1 = get_joker_counts(hand1[0])
    counts2 = get_joker_counts(hand2[0])

    # Compare the counts...
    for c1, c2 in zip(counts1, counts2): 
        if c1 > c2: 
            return 1
        elif c2 > c1: 
            return -1
        
    # No returns? Let's return based on the first card!
    return strongest_suit(hand1, hand2, cards_ordered)

def determine_total_winnings(hands):
    total_winnings = 0
    for i, (hand, bid) in enumerate(hands): 
        total_winnings += int(bid)*(i+1)
    return total_winnings

hands.sort(key=cmp_to_key(part1_compare_hands))
print('Part 1', determine_total_winnings(hands))

hands.sort(key=cmp_to_key(part2_compare_hands))
print('Part 2', determine_total_winnings(hands))
```
